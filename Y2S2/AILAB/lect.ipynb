# Differences between cv2 and np operations

```python
img = cv2.imread("Lenna.png")
M = np.full(img.shape, 10, dtype=np.int8)

added_img = cv2.add(img, M, dtype=1)

cv2.imshow("black", added_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
rect1 = np.zeros((500,500), dtype= "uint8")
rect2 = np.zeros((500,500), dtype= "uint8")

cv2.rectangle(rect1, (200,100), (300,300), (255,255,255), -1)
cv2.rectangle(rect2, (100,50), (250,250), (255,255,255), -1)

out = cv2.bitwise_xor(rect1, rect2)

cv2.imshow("out", out)

cv2.waitKey()
cv2.destroyAllWindows()
```

## Masks

```python
img = cv2.imread("Lenna.png")
mask = np.zeros(img.shape, dtype = np.int8)

cv2.rectangle(mask, (40,40), (100,100), (255,255,255), -1)

print(img.shape, mask.shape)
result = cv2.bitwise_and(img, mask)

cv2.imshow("out", result)

cv2.waitKey()
cv2.destroyAllWindows()
```

## Plotting the distribution of colors

```python
import matplotlib.pyplot as plt

img = cv2.imread("blorg.jpg")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_eq = cv2.equalizeHist(gray)
hist = cv2.calcHist([gray], [0], None, [255], [0,255]) # None for the mask, we are interested in the whole image
hist_eq = cv2.calcHist([gray_eq], [0], None, [255], [0,255])
plt.plot(hist)
plt.plot(hist_eq)
plt.show()

```

```python
color = ("b", "g", "r")
for i, col in enumerate(color):
    hist = cv2.calcHist([img], [i], None, [256], [0,255])
    plt.plot(hist, color=col)

plt.show()
```
